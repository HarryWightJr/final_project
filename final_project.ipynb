```python
# Add the Pandas dependency.
import pandas as pd
import os
```

```python
# Files to load
crime_2019 = "Resources/crime_2019.csv"
crime_2020 = "Resources/crime_2020.csv"
city_sales = "Resources/Sale_Prices_City.csv"
```

```python
crime_2019_df = pd.read_csv(crime_2019)
crime_2020_df = pd.read_csv(crime_2020)
city_sales_df = pd.read_csv(city_sales)
```

```python
# Combine the data into a single dataset.
partial_df = pd.merge(crime_2019_df, crime_2020_df, on=["CountyID"])
partial_df.head()
```

```python
complete_df = pd.merge(partial_df, city_sales_df, on=["CountyID"])
complete_df.head()
```

```python
renamed_df = complete_df.rename(columns={'City_x': 'City', 'State_x': 'State', "CountyID": "County ID", "Population_x": "Population", "Violent\ncrime": "Violent Crime", "Murder and\nnonnegligent\nmanslaughter": "Murder and\ Nonnegligent\ Manslaughter", "Rape_x": "Rape", "Robbery_x": "Robbery", "Aggravated\nassault": "Aggravated Assault", "Burglary_y": "Burglary", "Arson_y": "Arson"})  
```
