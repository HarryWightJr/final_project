```python
# Add the Pandas dependency.
import pandas as pd
import os
```

```python
# Files to load
crime_2019 = "Resources/crime_2019.csv"
crime_2020 = "Resources/crime_2020.csv"
city_sales = "Resources/Sale_Prices_City.csv"
income = "Resources/Income.csv"
```

```python
crime_2019_df = pd.read_csv(crime_2019)
crime_2020_df = pd.read_csv(crime_2020)
city_sales_df = pd.read_csv(city_sales)
income_df = pd.read_csv(income)
```

```python
crime_2019_df.head()
```

```python
dropped_crime_2019_df = crime_2019_df.drop(columns=["City", "State","City and State"])
sorted_crime_2019_df = dropped_crime_2019_df.sort_values("Population 2019")
final_crime_2019_df = sorted_crime_2019_df.drop_duplicates(subset=['CountyID'], keep='last')
final_crime_2019_df.head()
```

```python
final_crime_2019_df.count()
```

```python
crime_2020_df.head()
```

```python
dropped_crime_2020_df = crime_2020_df.drop(columns=["City", "State","City and State"])
sorted_crime_2020_df = dropped_crime_2020_df.sort_values("Population 2020")
final_crime_2020_df = sorted_crime_2020_df.drop_duplicates(subset=['CountyID'], keep='last')
final_crime_2020_df.head()
```

```python
final_crime_2020_df.count()
```

```python
city_sales_df.head()
```

```python
dropped_city_sales_df = city_sales_df.drop(columns=["RegionName", "StateName"])
final_city_sales_df = dropped_city_sales_df.drop_duplicates(subset=['CountyID'], keep='last')
final_city_sales_df.head()
```

```python
final_city_sales_df.count()
```

```python
income_df.head()
```

```python
dropped_income_df = income_df.drop(columns=["County", "State","County, State"])
final_income_df = dropped_income_df.drop_duplicates(subset=['CountyID'], keep='last')
final_income_df.head()
```

```python
final_income_df.count()
```

```python
# Combine the data into a single dataset.
partial_1_df = pd.merge(final_city_sales_df, final_crime_2019_df, how="left", on=["CountyID"])
partial_1_df.head()
```

```python
partial_2_df = pd.merge(partial_1_df, final_crime_2020_df, how="left", on=["CountyID"])
partial_2_df.head()
```

```python
complete_df = pd.merge(partial_2_df, final_income_df, how="left", on=["CountyID"])
complete_df.head()
```

```python
complete_df.count()
```
